### Spliting the data into train and test

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.preprocess_data, df.target,test_size=.25,random_state=42)

print(f"x_train shape :{x_train.shape}")
print(f"y_train shape :{y_train.shape}")
print(f"x_test shape :{x_test.shape}")
print(f"y_test shape :{x_test.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Assuming you have x_train and x_test as your text data
vocab_size = 5000  # Adjust this value based on your dataset

tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(x_train)

x_train_sequences = tokenizer.texts_to_sequences(x_train)
x_test_sequences = tokenizer.texts_to_sequences(x_test)

max_word_length = 80
x_train_padded = pad_sequences(x_train_sequences, maxlen=max_word_length, padding='post', truncating='post')
x_test_padded = pad_sequences(x_test_sequences, maxlen=max_word_length, padding='post', truncating='post')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNNCell, Dense
```

## Let's Build the RNN model

```python
embedding_dim = 50
vocab_size = 8897
max_word_length = 80

model = Sequential()
model.add(Embedding(input_dim = vocab_size, output_dim = embedding_dim, input_length=max_word_length))
model.add(SimpleRNN(units=80,activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

## let' Compile the model and run

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

checkpoint = ModelCheckpoint("best_model.h5",
                             monitor='val_loss',
                             save_best_only=True,
                             mode='min',
                             verbose=1
                             )
early_stopping = EarlyStopping(monitor='val_loss',
                               patience=4,
                               mode='min',
                               restore_best_weights=True,
                               verbose=1
                               )

history = model.fit(x_train_padded, y_train, epochs=100, batch_size=32, validation_split=0.2,
                    callbacks=[checkpoint, early_stopping])
```

```python
import matplotlib.pyplot as plt

# Assuming you have already trained the model and have the 'history' object

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

### Let's Try Different Architecure (LSTM)

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(x_train)

x_train_sequences = tokenizer.texts_to_sequences(x_train)
x_test_sequences = tokenizer.texts_to_sequences(x_test)

max_word_length = 80
x_train_padded = pad_sequences(x_train_sequences, maxlen=max_word_length, padding='post', truncating='post')
x_test_padded = pad_sequences(x_test_sequences, maxlen=max_word_length, padding='post', truncating='post')
```

```python
from tensorflow.keras.layers import LSTM, Dropout
output_dim=100
input_dim=80
model1 = Sequential()
model1.add(Embedding(input_dim=input_dim, output_dim=output_dim, input_length=max_word_length))
model1.add(LSTM(units=80, activation='tanh', return_sequences=True))
model1.add(Dropout(0.2))  
model1.add(LSTM(units=80, activation='tanh'))
model1.add(Dense(units=1, activation='sigmoid'))

model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model1.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

checkpoint = ModelCheckpoint("best_model_LSTM.h5",
                             monitor='val_loss',
                             save_best_only=True,
                             mode='min',
                             verbose=1
                             )
early_stopping = EarlyStopping(monitor='val_loss',
                               patience=4,
                               mode='min',
                               restore_best_weights=True,
                               verbose=1
                               )

history = model1.fit(x_train_padded, y_train, epochs=100, batch_size=32, validation_split=0.2,
                    callbacks=[checkpoint, early_stopping])
```

```python
import matplotlib.pyplot as plt

# Assuming you have already trained the model and have the 'history' object

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
class Square:
    def draw(self):
        print(f'Inside Square: :draw()')

    def resize(self):
        print(f'Inside Square::resize()')

class Circle:
    def draw(self):
        print(f'Inside Circle::draw()')

    def resize(self):
        print(f'Inside Circle::resize()')
class ShapeManager:
  def __init__(self, shapes):
    self._shapes = shapes

  def manage(self):
    for shape in self._shapes:
      shape.draw()
      shape.resize()

if __name__ == '__main__':
  shapes = [Square(), Square(), Circle(), Square(), Circle(), Circle(), Square(), Circle()]
  shape_manager = ShapeManager(shapes)
  shape_manager.manage()
```
